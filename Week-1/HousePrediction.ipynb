```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("USA_Housing.csv")
df.columns = ["Income", "House Age", "Number of Rooms", "Number of Bedrooms", "Population", "Price", "Address"]

df["Address"] = df["Address"].str.replace('\n', ' ')
df['Price'] = df['Price'].apply(lambda x: '{:.2f}'.format(x))

df
```

```python
# Feature'lar
X = df[["Income", "House Age", "Number of Rooms", "Number of Bedrooms", "Population"]]
# Target'lar
y = df["Price"]

from sklearn.model_selection import train_test_split

# Dataset'i X ve y'leri train ve test dataset'lerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=101)

from sklearn.linear_model import LinearRegression

# Train set'lere Linear Regression uygulama
lm = LinearRegression()
lm.fit(X_train, y_train)

y_pred = lm.predict(X_test)
y_pred

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)

coeff_df = pd.DataFrame(lm.coef_, X.columns, columns = ["Coefficient"])

# Buradaki çıkan sayılar mesela: 1 unit artarsa income 21 $ artacak demek. Formüldeki feature'ların coefficient'ları
coeff_df

predictions = lm.predict(X_test)

plt.scatter(y_test, predictions)
```
