```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import add_dummy_feature
from sklearn.linear_model import LinearRegression

np.random.seed(42)
m = 100 # number of instances
X = 2 * np.random.rand(m, 1) # column vector, 0-2 arasında sayı üret
y = 4 + 3 * X + np.random.rand(m, 1) # 4 + 3 * x fonksiyonuna gürültü ekleme

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b. ")
plt.xlabel("$x_l$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
X_b = add_dummy_feature(X)
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y # Normal Equation ile MSE cost function'ını minimize eden Theta değeri

theta_best # İlk baştaki fonskiyonda theta değeri 4 idi ama şu an 4.51 bulundu

X_new = np.array([[0], [2]])
X_new_b = add_dummy_feature(X_new) # add x0 = 1 to each instance
y_predict = X_new_b @ theta_best
y_predict
```

```python
plt.plot(X_new, y_predict, "r-", label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_l$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.legend(loc="upper left")
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_

lin_reg.predict(X_new)
```
