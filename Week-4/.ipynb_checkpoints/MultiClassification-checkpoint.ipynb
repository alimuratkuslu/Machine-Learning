```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
import seaborn as sns

# Defining the columns and changing the classes from string to integers
cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'target']
iris_dict = {'Iris-setosa': 0.0, 'Iris-versicolor': 1.0, 'Iris-virginica': 2.0}

df = pd.read_csv('iris.data', names=cols, header=None)

df["target"] = df["target"].replace(iris_dict)
df
```

```python
# Shuffling the dataset

df = df.sample(frac=1)
```

```python
# Train, test split

train_set, test_set = train_test_split(df, test_size=0.4)

# Splitting test_set into %50 valid and %50 test

test_dataset, valid_set = train_test_split(test_set, test_size=0.5)

train_set
```
