```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

X = np.arange(10)
Y = (X-5)**2

plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
x = 0
lr = 0.1
error = []

plt.plot(X, Y)

for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x, y)
```
